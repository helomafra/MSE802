```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
import math
from calculator import Calculator

# Configure graph style
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['font.size'] = 12
```

```python
# Instantiate the calculator
calc = Calculator()
print("Complex number calculator created successfully!")
```

```python
# Getting the complex numbers from the user
real1 = float(input("Real part of z1: "))
imag1 = float(input("Imaginary part of z1: "))
z1 = calc.create(real1, imag1)

real2 = float(input("Real part of z2: "))
imag2 = float(input("Imaginary part of z2: "))
z2 = calc.create(real2, imag2)
```

```python
print(f"z₁ = {z1}")
print(f"z₂ = {z2}")
```

```python
# Example 1: Addition
result_sum = calc.add(z1, z2)
print(f"Result: {z1} + {z2} = {result_sum}")

# Show result
calc.show(result_sum, "Result")

# Plot the vectors
calc.plot_vectors(z1, z2, result_sum, f"({z1}) + ({z2})")
```

```python
# Example 2: Subtraction
result_sub = calc.subtract(z1, z2)
print(f"\nSubtraction: {z1} - {z2} = {result_sub}")

# Show result
calc.show(result_sub, "Subtraction result")

# Plot subtraction
calc.plot_vectors(z1, z2, result_sub, f"({z1}) - ({z2})")
```

```python
# Example 3: Multiplication
result_mult = calc.multiply(z1, z2)
print(f"\nMultiplication: {z1} × {z2} = {result_mult}")

# Show result
calc.show(result_mult, "Multiplication result")

# Plot multiplication
calc.plot_vectors(z1, z2, result_mult, f"({z1}) × ({z2})")
```

```python
# Example 4: Division
result_div = calc.divide(z1, z2)
print(f"\nDivision: {z1} ÷ {z2} = {result_div}")

# Show result
calc.show(result_div, "Division result")

# Plot division
calc.plot_vectors(z1, z2, result_div, f"({z1}) ÷ ({z2})")
```
